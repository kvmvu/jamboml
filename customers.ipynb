```python
import pandas as pd
import numpy as np

from keys import conn_str
from sqlalchemy import create_engine, exc
```

```python
df = pd.read_sql_table('customers_raw', conn_str)
df.tail()
```

```python
df2 = pd.read_csv('mpesa_kopokopo.csv', header=1)
df2.head()
```

```python
df2.info()
```

```python
df2.Phone = df2['Phone'].astype(str)
```

```python
df2['Data.Code'] = df2['Phone'].str.slice(start=5)
df2.head()
```

```python
new_df = pd.merge(df2,df, how='inner')
new_df
```

```python
new_df = new_df.drop(columns=['Transaction ID', 'Till No.', 'Fees', 'MM System', 'Status', 'index', 'Data.City',
                              'Data.Email', 'Data.Mobile', 'Data.Phone', 'Data.Name', 'TotalCount'])

new_df
```

```python
# sorting out datetime values
new_df['Date'] = pd.to_datetime(new_df['Date'])
new_df = new_df.set_index('Date')

new_df['pay_year'] = new_df.index.year
new_df['pay_month'] = new_df.index.month
new_df['pay_day'] = new_df.index.day
new_df['pay_dayOfWeek'] = new_df.index.weekday
new_df['pay_weekOfMonth'] = new_df.index.week
new_df['pay_hour'] = new_df.index.hour
new_df['pay_minute'] = new_df.index.minute
new_df.head()

new_df = new_df.reset_index()
new_df
```

```python
engine = create_engine(conn_str, echo=True)

try:
    new_df.to_sql('customers_clean', con=engine, if_exists='replace', index=False)
except exc.IntegrityError:
    pass
e = engine.execute("SELECT * FROM customers_clean").fetchall()
print(e)
```
